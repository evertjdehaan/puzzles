Toch maar even proberen welke letters in de versleutelde tekst dezelfde letters in de originele tekst zouden kunnen zijn.

```python
import re
from itertools import permutations

# Get all unique letters
letters = set(message)

options = []

# Find the letters which occur in an alternating fashion (might indicate the same letter)
for (letter_a, letter_b) in permutations(letters, 2):
  # Remove all letters not equal to the two letters of interest
  only_a_and_b = re.sub(r'[^'+letter_a+letter_b+']', '', message)
  # Check if the three letters really alternate as abab...
  if (only_a_and_b.find(letter_a*2) == -1) and (only_a_and_b.find(letter_b*2) == -1):
    options.append(re.sub(r'[^'+letter_a+letter_b+']', '_', message))

for option in options:
  print(option)
```

```python
import re
from itertools import permutations

# Get all unique letters
letters = set(message)

options = []

# Find the letters which occur in an alternating fashion (might indicate the same letter)
for (letter_a, letter_b, letter_c) in permutations(letters, 3):
  # Remove all letters not equal to the three letters of interest
  only_a_b_and_c = re.sub(r'[^'+letter_a+letter_b+letter_c+']', '', message)
  # Check if the three letters really alternate as abcabc...
  if ((only_a_b_and_c.find(letter_a*2) == -1) and 
      (only_a_b_and_c.find(letter_b*2) == -1) and
      (only_a_b_and_c.find(letter_c*2) == -1) and
      (only_a_b_and_c.find(letter_a+letter_b+letter_a) == -1) and
      (only_a_b_and_c.find(letter_a+letter_c+letter_a) == -1) and
      (only_a_b_and_c.find(letter_b+letter_a+letter_b) == -1) and
      (only_a_b_and_c.find(letter_b+letter_c+letter_b) == -1) and
      (only_a_b_and_c.find(letter_c+letter_a+letter_c) == -1) and
      (only_a_b_and_c.find(letter_c+letter_b+letter_c) == -1)):
    options.append(re.sub(r'[^'+letter_a+letter_b+letter_c+']', '_', message))
  
for option in options:
  print(option)
```
