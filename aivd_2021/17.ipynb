```python
message = '.......------...-..-..-....-.......-----.---.-.-.-.--...-...-...-.--.....---..---...........--..---.---..--..-....-.--.-.--.....----.-.-----..-.-..-...----...--..-..--.-.....-..--.'

letters = ['.-', '-...', '-.-.', '-..', '.', '..-.', '--.',
           '....', '..', '.---', '-.-', '.-..', '--', '-.',
           '---', '.--.', '--.-', '.-.', '...', '-', '..-',
           '...-', '.--', '-..-', '-.--', '--..']
numbers = ['-----', '.----', '..---', '...--', '....-',
           '.....', '-....', '--...', '---..', '----.']

def _get_char(msg, start, end):
  if (end-start) < 5:
    pos = letters.index(msg[start:end])
    return chr(65+pos)
  else:
    pos = numbers.index(msg[start:end])
    return str(pos)

decoded = 'HIJ OVERLEED 50 JAAR GELEDEN ZIJ 25 HUN JAZZDUETTEN ZIJN NOG ALTIJD'
start = 0
for char in decoded.replace(' ',''):
  pos = ord(char)-65
  if pos >= 0: # it's a letter
    start += len(letters[ord(char)-65])
  else: # it's a number
    start += 5

include_numbers = True
if include_numbers:
  max_token_length = 5
else:
  max_token_length = 4

search_len = 6
options = []
for i in range(1,max_token_length+1):
  for j in range(1,max_token_length+1):
    for k in range(1,max_token_length+1):
      for l in range(1,max_token_length+1):
        for m in range(1,max_token_length+1):
          for n in range(1,max_token_length+1):
            option = ''
            if search_len >= 1:
              try:
                option += _get_char(message, start, start+i)
              except ValueError:
                continue

            if search_len >= 2:
              try:
                option += _get_char(message, start+i, start+i+j)
              except ValueError:
                continue
            
            if search_len >= 3:
              try:
                option += _get_char(message, start+i+j, start+i+j+k)
              except ValueError:
                continue

            if search_len >= 4:
              try:
                option += _get_char(message, start+i+j+k, start+i+j+k+l)
              except ValueError:
                continue

            if search_len >= 5:
              try:
                option += _get_char(message, start+i+j+k+l, start+i+j+k+l+m)
              except ValueError:
                continue

            if search_len == 6:
              try:
                option += _get_char(message, start+i+j+k+l+m, start+i+j+k+l+m+n)
              except ValueError:
                continue
          
            if len(option) == search_len: # if this is not met, invalid letters were found
              options.append(option)

for option in sorted(set(options)):
  print(option)
```
